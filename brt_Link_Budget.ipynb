# Input Parameters

```python
#@title ### GS Parameters
lat = -54.5104         # Latitude (°)
lon = -67.11556        # Longitude (°)
Gsh = 0.172            # Earth station altitude (km)
D = 13.5               # Earth station antenna diameter (m)
fup = 2.0769396        # Up Link frequency (GHz)
E_deg = 5.0            # Elevation angle (deg)
to = 45.0              # Tilt polarization angle (°) for Circular polarization
eta = 0.5              # Earth station antenna efficiency
p = 0.01               # Percentage of the time for rain calculation
GsEIRP = 61            # EIRP Ground Station
GsAxRt = [1, 1.5, 0.5] # Axial Ratio (dB) [NOM ADV FAV]
GsPoLs = 0.03          # GS Pointing Losses (dB)
GsGoverT = 24          # GS G over T
```

```python
#@title ### SC Parameters
Sch = 639              # Satellite Altitude (km) [638 651 619] (NOM, ADV, FAV)
fdw = 2.2555           # Down Link frequency (GHz)
ScTxPwr = 37           # S/C Tx Power (dBm)
ScAntGdw = -7.8        # S/C Antenna Gain @ DownLink Freq (dBi)
TmMI = 1.4             # Telemetry modulation index (sine)
RnMI = 0.7             # Ranging modulation index
```

```python
#@title ### Other reqs and values
# Values from UPLINK BUDGET
RnNoBw = 60.79         # Ranging noise BW (dBHz)
ScRx = -83.86          # Rx power at receiver input (dBm)
ScRxSnNo = 85.24       # Ranging signal-to-noise power ratio at the output of the transponder filter.

# Values from UPLINK BUDGET & Calculation of MI
TcRnLs = 10.18         # Pérdida de Modulación RNG
TcLs =  4.93           # Pérdida de Comando

# Carrier Recovery input data
DemoTecLs = 0          # (dB) Demodulator technological Loss
PLLBw_hz = 300         # (Hz) PLL Bandwidth (2*BL)
CNreq = 17             # (dB) required C/N

# LBR TM RECOVERY input data
RcvrDemoTecLs = 3      # (dB)
EbN0req = 10.2         # (dB)
TMdatarate_bps = 4000  # (bits/s)

# RANGING RECOVERY input data
RNGDemoTecLs = 1       # (dB) Demodulator technological Loss
RNGPLLBw_hz = 3        # (Hz) PLL Bandwidth (2*BL)
RNGSNreq = 19          # (dB) required S/N

# General Parameters
PolLs = 0.06           # Polarization Loses
```

# Cálculos

\begin{equation}
E = E_{deg} \cdot \frac{π}{180}
\tag{1}
\end{equation}

<br>

\begin{equation}
ScEIRP = ScTxPwr + ScAntGdw - 30
\tag{2}
\end{equation}

<br>

\begin{equation}
\Lambda_{Dw} = \frac{3 \times 10^8}{fdw \times 10^9}
\tag{3}
\end{equation}

<br>

\begin{equation}
ScAOV = \arcsin \left( \frac{6378.14}{6378.14 + \text{Sch} - {Gsh}} \cdot \sin \left( \frac{\pi}{2} + \frac{\pi \cdot E_{deg}}{180} \right) \right) \cdot \frac{180}{\pi}
\tag{4}
\end{equation}



```python
import math
# from scipy.special import besselj
from scipy.special import jn

E = E_deg * (math.pi / 180) # eq1

# Intermediate parameters computation
hr = propa64.rain_height(c_double(lat), c_double(lon))
R001 = propa64.rain_intensity(c_double(lat), c_double(lon), c_double(0.01))
Temp = propa64.temperature(c_double(lat), c_double(lon))
MeanRTemp = propa64.TMR(c_double(Temp))
WVC = propa64.SWVD(c_double(lat), c_double(lon))

if p >= 1:
    Iwvc = propa64.IWVC(c_double(lat), c_double(lon), c_double(p))
    TCC = propa64.LWCC(c_double(lat), c_double(lon), c_double(p))
else:
    Iwvc = propa64.IWVC(c_double(lat), c_double(lon), c_double(1))
    TCC = propa64.LWCC(c_double(lat), c_double(lon), c_double(1))

Nwet = propa64.NWET(c_double(lat), c_double(lon))

# Final parameters computation - Downlink Budget
ScEIRP = ScTxPwr + ScAntGdw - 30 # eq2
LambdaDw = 3E8 / (fdw * 1E9)  # eq3
ScAOV = math.asin((6378.14 / (6378.14 + Sch - Gsh)) * math.sin(math.pi / 2 + math.pi * E_deg / 180)) * (180 / math.pi)  # eq4
# SlR = math.sqrt(math.pow(6.37814, 2) + math.pow((6.37814 + ((Sch - Gsh) / 1000)), 2) - 2 * 6.37814 * (6.37814 + ((Sch - Gsh) / 1000)) * math.cos(math.pi / 2 - (math.pi * E_deg / 180 + math.asin(6.37814 / (6.37814 + ((Sch - Gsh) / 1000)) * math.sin(math.pi / 2 + math.pi * E_deg / 180)))) * 1000
SlR = math.sqrt(math.pow(6.37814, 2) +
                math.pow((6.37814 + ((Sch - Gsh) / 1000)), 2) -
                2 * 6.37814 * (6.37814 + ((Sch - Gsh) / 1000)) *
                math.cos(math.pi / 2 - (math.pi * E_deg / 180 + math.asin(6.37814 / (6.37814 + ((Sch - Gsh) / 1000)))) *
                math.sin(math.pi / 2 + math.pi * E_deg / 180))) * 1000
FSLdw = 32.44 + 20 * math.log10(fdw * 1E3) + 20 * math.log10(SlR)

# PROPA
# Calcular las pérdidas debido a las diferentes condiciones atmosféricas
CloudLssdw = propa64.cloud_attenuation(fdw, E, TCC)
GasLssdw = propa64.gaseous_attenuation(fdw, E, Temp, WVC)
RainAttdw = propa64.rain_attenuation(lat, fdw, E, p, Gsh, hr, R001, to)
TropoLssdw = propa64.scintillation(Nwet, fdw, E, p, Gsh, eta, D)

# Calcular las pérdidas atmosféricas totales
AtmLss = GasLssdw + math.sqrt((CloudLssdw + RainAttdw) ** 2 + TropoLssdw ** 2)

GsIPFD = ScEIRP - GsPoLs - 10.99 - 20 * math.log10(SlR * 1000)
GsIPFDmx = GsIPFD + AtmLss
SnNo = ScEIRP - FSLdw - AtmLss - PolLs - GsPoLs + GsGoverT + 228.6

# Modulation Index Calculations
No = ScRx - ScRxSnNo
NoRb = No + RnNoBw
alphaR = math.pow(10, (ScRx - TcRnLs - NoRb) / 10)
alphaC = math.pow(10, (ScRx - TcLs - NoRb) / 10)
ENI = RnMI / math.sqrt(2 * (alphaC + alphaR + 1))
ERI = RnMI * math.sqrt(alphaR / (alphaC + alphaR + 1))
TcEI = RnMI * math.sqrt(alphaC / (alphaC + alphaR + 1))

Csup = -(20 * (math.log10(abs(jn(0, TmMI))))) - \
       (20 * (math.log10(abs(jn(0, TcEI))))) - \
       (20 * (math.log10(abs(jn(0, ENI))))) - \
       (20 * (math.log10(abs(jn(0, ERI))))) - \
       10 * math.log10(math.exp(-math.pow(ENI, 2)))

PLLBw = 10 * math.log10(PLLBw_hz)
CNavalb = SnNo - Csup - DemoTecLs - PLLBw
Cmarg = CNavalb - CNreq

ModLs = -(20 * (math.log10((jn(1, TmMI)))))    \
        -(20 * (math.log10(abs(jn(0, TcEI))))) \
        -(20 * (math.log10(abs(jn(0, ENI)))))  \
        -(20 * (math.log10(abs(jn(0, ERI)))))  \
        -10 * math.log10(math.exp(-math.pow(ENI, 2)))

TMdatarate = 10 * math.log10(TMdatarate_bps)
EbN0avalb = SnNo - ModLs - RcvrDemoTecLs - TMdatarate
LBRTMmarg = EbN0avalb - EbN0req

RNGmodLs = -(20 * (math.log10((jn(1, ERI))))) \
           -(20 * (math.log10(abs(jn(0, TmMI))))) \
           -(20 * (math.log10(abs(jn(0, ENI))))) \
           -(20 * (math.log10(abs(jn(0, TcEI))))) \
           -10 * math.log10(math.exp(-math.pow(ENI, 2)))

RNGPLLBw = 10 * math.log10(RNGPLLBw_hz)
RNGSNavalb = SnNo - RNGmodLs - RNGDemoTecLs - RNGPLLBw
RNGSNmarg = RNGSNavalb - RNGSNreq

# Final parameters computation - Uplink Budget
LambdaUp = 3E8 / (fup * 1E9)
FSLup = 32.44 + 20 * math.log10(fup * 1E3) + 20 * math.log10(SlR)

# PROPA
CloudLssup = propa64.cloud_attenuation(c_double(fup), c_double(E), c_double(TCC))  # Cloud Losses | C16
GasLssup = propa64.gaseous_attenuation(c_double(fup), c_double(E), c_double(Temp), c_double(WVC))  # Gaseosous Losses | C17
RainAttup = propa64.rain_attenuation(c_double(lat), c_double(fup), c_double(E), c_double(p), c_double(Gsh), c_double(hr), c_double(R001), c_double(to))  # Rain Attenuation | C18
TropoLssup = propa64.scintillation(c_double(Nwet), c_double(fup), c_double(E), c_double(p), c_double(Gsh), c_double(eta), c_double(D))  # Tropospheric scentillation losses | C19

AtmLssup = GasLssup + math.sqrt(math.pow((CloudLssup + RainAttup), 2) + math.pow(TropoLssup, 2))  # Overall atmospheric losses =C17+SQRT((POWER((C18+C16);2)+POWER(C19;2)))

```

```python
#@title No es necesario correr esta celda
lat = 46.217          #// Latitude (°)
lon = 6.12            #// Logitude (°)

hs = 0.8140           #// Earth station altitude (km)
E = 33.0*math.pi/180  #// Link elevation angle (°)

f1 = 12.0             #// Link frequency (GHz)
f2 = 30               #// Second link frequency f2 > f1 (for Frequency scaling) (GHz)
to = 45.0             #// Tilt polarization angle (°)

D = 1.0               #// Earth station antenna diameter (m)
eta = 0.5             #// Earth station antenna efficiency

p = 1                  #// Percentage of the time

#// Intermediate parameters computation
hr = propa64.rain_height(lat, lon)  # Sin c_double aquí
R001 = propa64.rain_intensity(lat, lon, 0.01)  # Sin c_double aquí
Temp = propa64.temperature(lat, lon)  # Sin c_double aquí
ro = propa64.SWVD(lat, lon)  # Sin c_double aquí
if p >= 1:
    WVC = propa64.IWVC(lat, lon, p)  # Sin c_double aquí
    LWC = propa64.LWCC(lat, lon, p)  # Sin c_double aquí
else:
    WVC = propa64.IWVC(lat, lon, 1)  # Sin c_double aquí
    LWC = propa64.LWCC(lat, lon, 1)  # Sin c_double aquí

Nwet = propa64.NWET(lat, lon)  # Sin c_double aquí

print("\n==========================\n")
print("Link frequency %.2lf GHz \n" % f1)
print("==========================\n")


#// GASEOUS ATTENUATION
Agaseous = propa64.gaseous_attenuation(f1, E, Temp, ro)
print("\nGaseous attenuation = %.2lf dB\n" % Agaseous)

#// GASEOUS ATTENUATION EXCEEDED FOR p% OF THE TIME
Agaseous_exc = propa64.gaseous_attenuation_exc(f1, E, Temp, WVC)
print("\nGaseous attenuation exceeded for %.2lf %% of the time = %.2lf dB\n" % (p, Agaseous_exc))

#// CLOUD ATTENUATION EXCEEDED FOR p% OF THE TIME
Aclouds = propa64.cloud_attenuation(f1, E, LWC)
print("\nClouds attenuation exceeded for %.2lf %% of the time = %.2lf dB\n" % (p, Aclouds))

#// IMPAIRMENT DUE TO SCINTILLATION EXCEEDED FOR p% OF THE TIME
Iscint = propa64.scintillation(Nwet, f1, E, p, hs, eta, D)
print("\nScintillation impairment exceeded for %.2lf %% of the time = %.2lf dB\n" % (p, Iscint))

#// RAIN ATTENUATION EXCEEDED FOR p% OF THE TIME
Arain = propa64.rain_attenuation(lat, f1, E, p, hs, hr, R001, to)
print("\nRain attenuation exceeded for %.2lf %% of the time = %.2lf dB\n" % (p, Arain))

#// TOTAL ATTENUATION
Atotal = Agaseous_exc + math.sqrt((Arain + Aclouds)*(Arain + Aclouds) + Iscint*Iscint)
print("\nTotal attenuation exceeded for %.2lf %% of the time = %.2lf dB\n" % (p, Atotal))

#// SKY NOISE TEMPERATURE FOR A GIVEN TOTAL ATTENUATION
MeanRTemp = propa64.TMR(Temp)
NoiseTemp = propa64.noise_temperature(Atotal, MeanRTemp)
print("\nSky noise temperature for the given total attenuation %.2lf dB = %.2lf K\n" % (Atotal, NoiseTemp))

#// CROSS POLAR DISCRIMINATION EXCEEDED FOR p% OF THE TIME
XPOL = propa64.XPD(Arain, to, f1, E, p)
print("\nCross polar discrimination exceeded for %.2lf %% of the time = %.2lf dB\n" % (p, XPOL))

print("\n====================================\n")
print("Link frequency scaling to %.2lf GHz \n" % f2)
print("====================================\n")

#// RAIN ATTENUATION ON FREQ 2 FROM FREQ 1 (using frequency scaling)
Arain2 = propa64.EFSR(f1, f2, Arain)
print("\nRain attenuation exceeded on freq. %.2lf GHz based on freq. %.2lf GHz = %.2lf dB\n" % (f2, f1, Arain2))
```
